## 배낭문제 분기한정 알고리즘
* 분기 한정 알고리즘으로 해를 찾으려면 새 노드를 만들 때 노드의 한정 값을 계산해야  
* 노드의 한정 값(bound)은 그 노드가 얼마나 해에 가까운지를 알 수 있는 척도
* 최대 가치를 찾는 문제이므로 높은 한정 값을 가진 노드들을 우선하여 탐색, 즉 최선 우선(Best-First) 탐색
* bound(한정값) = [현재 담긴 물건의 가치] + [배낭의 남은 부분에 담을 물건의 가치]
> * 배낭 문제는 배낭의 용량을 초과하지 않으면서 물건들을 배낭에 담아 최대 가치를 얻는 것이 목적
> * 따라서 노드의 한정 값이 그 노드로부터 탐색하면 얻을 수 있는 최대 가치(예측값)를 가질 때 상태 공간 트리에서 가장 효율적인 탐색이 가능
